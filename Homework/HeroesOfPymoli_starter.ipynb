```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file = "/Users/bryanwood/Desktop/Pandas_Homework/Homework/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file)
purchase_data.head(10)
```

## Player Count

* Display the total number of players


```python
#Add column to assign each row a value 
purchase_data["Player"] = 1

#Count total players and store it into a variable
total_players = purchase_data["Player"].count()

#Print total players
print(f"Total Players: {total_players}")
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Calculate unique items and store into variable
unique_items = purchase_data["Item ID"].nunique()

#Calculate average price and store into variable
average_price = purchase_data["Price"].mean()

#Calculate number of purchases and store into variable
num_purchases = purchase_data["Purchase ID"].nunique()

#Calculate total revenue and store into variable
total_revenue = purchase_data["Price"].sum()

#Create Summary data frame to hold results
purchase_analysis = pd.DataFrame({
                                    "Number of Unique Items": [unique_items],
                                    "Average Price": [average_price],
                                    "Number of Purchases": [num_purchases],
                                    "Total Revenue": [total_revenue]
                                })
#Format Average Price and Total Revenue data to currency and 2 decimal points
purchase_analysis["Average Price"] = purchase_analysis["Average Price"].map("${:.2f}".format)
purchase_analysis["Total Revenue"] = purchase_analysis["Total Revenue"].map("${:.2f}".format)

#Display summary data frame
purchase_analysis.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Grouping the DataFrame by "Gender"
gender_group = purchase_data.groupby("Gender")

#finding total gender count
gender_count = gender_group["Player"].sum()

#Calculate gender percentage
gender_percentage = ((gender_count/total_players)*100).map("{:.2f}%".format)

# Creating a new DataFrame using both percentage and count
gender_summary_table = pd.DataFrame({"Count": gender_count,
                                     "Percentage": gender_percentage})
gender_summary_table.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
#Calculate average purchase price
gender_avg_price = gender_group["Price"].mean().map("${:.2f}".format)

#Calculate total purchase value
gender_total_purchase = gender_group["Price"].sum().map("${:.2f}".format)

#Calculate total per person


# Creating a new DataFrame 
purchase_analysis_table = pd.DataFrame({
                                    "Purchase Count": gender_count,
                                    "Average Purchase Price": gender_avg_price,
                                    "Total Purchase Value": gender_total_purchase,
                                    })
purchase_analysis_table.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Create the bins in which Data will be held
# Bins are <10, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39, 50
bins = [0, 9, 14, 19, 24, 29, 34, 39, 50]

# Create the names for the four bins
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#Add age group as column and establish bins
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], bins, labels=group_names) 

#Group by the established bins
group_demographics = purchase_data.groupby("Age Group")
group_demographics.count()

gender_age_count = group_demographics["Age"].count()
gender_age_percentage = ((gender_age_count / total_players) * 100).map("{:.2f}%".format)

# Creating a new DataFrame 
gender_demo_table = pd.DataFrame({
                                    "Count": gender_age_count,
                                    "Percentage": gender_age_percentage,
                                    })
gender_demo_table
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Grouping the DataFrame by "Gender"
gender_group = purchase_data.groupby("Age Group")

#finding total gender count
gender_count = gender_group["Player"].sum()

#Calculate gender percentage
gender_percentage = ((gender_count/total_players)*100).map("{:.2f}%".format)

# Creating a new DataFrame using both percentage and count
gender_summary_table = pd.DataFrame({"Count": gender_count,
                                     "Percentage": gender_percentage})
gender_summary_table
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Grouping the DataFrame by "Gender"
sn_group = purchase_data.groupby("SN")

#finding total SN purchase count
sn_count = sn_group["Player"].sum()

#Calculate average purchase price per SN
sn_avg_price = sn_group["Price"].mean().map("${:.2f}".format)

#Calculate total purchase value per SN
sn_total_purchase = sn_group["Price"].sum().map("${:.2f}".format)


# Creating a new DataFrame using both percentage and count
sn_summary_table = pd.DataFrame({
                                    "Purchase Count": sn_count,
                                    "Average Purchase Price": sn_avg_price,
                                    "Total Purchase Value": sn_total_purchase
                                    })

sn_summary_table = sn_summary_table.sort_values("Purchase Count", ascending=False)
sn_summary_table.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# pulling item id, item name and price columns
most_popular_df = purchase_data[["Item ID","Item Name","Price"]]

# # grouping by item id and item name
grouped_popular = most_popular_df.groupby(["Item ID","Item Name"])

# #finding total item purchase count
grouped_popular_count = grouped_popular["Item Name"].count()

# item_price = most_popular_df["Price"]

#Calculate total purchase value per item
grouped_pop_total_purchase = most_popular_df["Price"].sum()

# Creating a new DataFrame to show purchase count, item price and total purchase value
grouped_pop_summary_table = pd.DataFrame({
                                    "Purchase Count": grouped_popular_count,
#                                     "Item Price" : item_price,
                                    "Total Purchase Value": grouped_pop_total_purchase
                                    })
grouped_pop_summary_table.head()
```
